```python
cv2.imshow("image", img[:, :,  (2,1,0)])   # 2 is red, 1 is grreen, 0 is blue RGB
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("image", img[:, :,  (1,0,2)])   # 1 is gren, 0 is blue, 2 is red GBR
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# access the part of the image 

```python
img[320:400, 100:600] # syntax is y1:y2, x1:x2# x1 = 100, y1 = 320, x2 = 600, y2 = 400
```

```python
cv2.imshow("image", img[320:400, 100:600])  
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# show image row wise

```python
for i in range(img.shape[0]):
    cv2.imshow("image", img[0:i+1, :])
    
    if cv2.waitKey(27) & 0xff == ord('q'):
        break
        
cv2.destroyAllWindows()
```

# show image column wise

```python
for i in range(img.shape[1]):
    cv2.imshow("image", img[:, 0:i+1])
    
    if cv2.waitKey(27) & 0xff == ord('q'):
        break
        
cv2.destroyAllWindows()
```
